## __2.1 Case Folding__

Case Folding digunakan untuk bisa memgubah kata/kalimat menjadi lowecase, menghapus angka dan tanda baca, menghapus whitespace (karakter kosong) yang mana tidak perlukan dalam pemodelan NLP

```python
# Import library regex
import re

# Buat fungction langkah dari case folding dan filtering
def caseFolding(text):
  text = text.lower() # Mengubah teks menjadi lower case
  text = re.sub(r'https?://\S+|www\.\S+', '', text) # Menghapus URL
  text = re.sub(r'[0-9]+', '', text) # Menghapus angka 
  text = re.sub(r'[^\w\s]', '', text) # Menghapus karakter tanda baca
  text = text.strip()
  return text
```

```python
raw_sample = df_klasifikasi['data teks'].iloc[381]
case_folding = caseFolding(raw_sample)
```

```python
print('Raw data \t:', raw_sample)
print('Case folding \t', case_folding)
```

## __2.2 Word Normalization__

Word Normalization digunakan untuk mengubah kata/kalimat slang menjadi seperti kata bakunya.

```python
# Download corpus slang word

!wget https://raw.githubusercontent.com/ksnugroho/klasifikasi-spam-sms/master/data/key_norm.csv
```

```python
# Meload data key_norm.csv
key_norm = pd.read_csv('/content/key_norm.csv')

def text_normalize(text):
  text = ' '.join([key_norm[key_norm['singkat']==word]['hasil'].values[0] if (key_norm['singkat']==word).any() else word for word in text.split()])
  text = str.lower(text)
  return text
```

```python
raw_sample = df_klasifikasi['data teks'].iloc[381]
case_folding = caseFolding(raw_sample)
normalize = text_normalize(case_folding)
```

```python
print('Raw data \t:', raw_sample)
print('Case folding \t', case_folding)
print('text_normalize \t', normalize)
```

## __2.3 Stopword Removal__

Stopword Removal (Filtering) adalah tahap mengambil kata-kata penting dari hasil token dengan menggunakan algoritma stoplist (membuang kata kurang penting) atau wordlist (menyimpan kata penting).

```python
# Import library stopword

from nltk.corpus import stopwords

stopwords_ind = stopwords.words('indonesian')
```

```python
len(stopwords_ind)
```

Stopword Indonesia

```python
# Membuat fungsi stopword removal

stopwords_ind = stopwords_ind

def remove_stop_word(text):
  clean_word = []
  text = text.split()
  for word in text:
    if word not in stopwords_ind:
      clean_word.append(word)
  return " ".join(clean_word)
```

```python
raw_sample = df_klasifikasi['data teks'].iloc[381]
case_folding = caseFolding(raw_sample)
normalize = text_normalize(case_folding)
stopword_removal = remove_stop_word(normalize)
```

```python
print('Raw data \t:', raw_sample)
print('Case folding \t', case_folding)
print('text_normalize \t', normalize)
print('Stop word \t', stopword_removal)
```

## __2.4 Stemming__

Stemming adalah proses Mengubah kata ke bentuk dasarnya, namun bentuk dasar tersebut tidak berarti sama dengan akar kata (root word).

```python
from nltk.text import TextCollection
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()

#Di sini kita buat fungsi stemming

def stemming(text):
  text = stemmer.stem(text)
  return text
```

```python
raw_sample = df_klasifikasi['data teks'].iloc[381]
case_folding = caseFolding(raw_sample)
normalize = text_normalize(case_folding)
stopword_removal = remove_stop_word(normalize)
text_stemming = stemming(stopword_removal)
```

```python
print ('Raw data \t:',raw_sample)
print ('Case folding \t:', case_folding)
print ('text_normalize \t:', normalize)
print ('Stop word \t:', stopword_removal)
print ('Stemming \t:', text_stemming)
```

## __2.5 Pre-processing Pipeline__

```python
# Disini kita buat pipelinenya (menggabungkan semua fungsi di atas)

def text_preprocessing(text):
  text = caseFolding(text)
  text = text_normalize(text)
  text = remove_stop_word(text)
  text = stemming(text)
  return text
```

```python
#Time untuk mendapatkan waktu komputasi
%%time 
#proses menjalankan fungsi text_preprocessing pada data teks
df_klasifikasi['clean data teks'] = df_klasifikasi['data teks'].apply(text_preprocessing)
```

## __3.1 Feature Extraction (TF-ID)__

```python
from sklearn.feature_extraction.text import TfidfVectorizer

#Unigram
tf_idf = TfidfVectorizer(ngram_range=(1,1))
tf_idf.fit(X)

X_tf_idf = tf_idf.transform(X)
```

## __3.2 Feature Selection__

```python
X = np.array(data_tf_idf) #data_unigram
y = np.array(y)

X
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

# Proses Algoritma Chi-Square
chi2_features = SelectKBest(chi2, k=152) #k=152
X_kbest_features = chi2_features.fit_transform(X, y)

# Melihat data
print('Original features:', X.shape[1])
print('Selected features:', X_kbest_features.shape[1])
```

```python
# Membuat datframe nilai setiap feature dari chi-square

data_chi2 = pd.DataFrame(chi2_features.scores_, columns=['Nilai'])
data_chi2
```

```python
#Menampilkan fitur beserta nilainya
feature = tf_idf.get_feature_names_out()
data_chi2['Fitur'] = feature
data_chi2
```

```python
# Mengurutkan fitur berdasarkan nilai tertinggi
data_chi2.sort_values(by='Nilai', ascending=False)
```

```python
#Menampilkan mask pada feature yang terseleksi

#False berarti fitur tidak terpilih dan True berarti fitur terpilih
mask = chi2_features.get_support()
mask
```

```python
#Menampilkan fitur-fitur terpilih berdasarkan mask atau nilai tertinggi

new_features = []

for bool, f in zip(mask,feature):
  if bool:
    new_features.append(f)
  selected_features = new_features
selected_features
```

```python
# Membuat vocabulary baru berdasarkan fitur yang terseleksi
#Ini digunakan untuk mengenerate fitur vector tf-idf

new_selected_features = {}

for (k,v) in tf_idf.vocabulary_.items():
    if k in selected_features:
        new_selected_features[k] = v

new_selected_features
```

```python
#Melihat jumlah fitur baru 
#Ini sesuai dengan nilai k pada chi-square
len(new_selected_features)
```

```python
#Memilih fitur terbaik dan memasukannya kedalam Dataframe
data_selected_features = pd.DataFrame(X_kbest_features,columns=selected_features)
data_selected_features
```

# __4. Modelling (Machine Learning)__

```python
#Import Library
from sklearn.model_selection import train_test_split

#Algoritme

#Algoritme Support Vector Machine
from sklearn import svm
#Naive Bayes
from sklearn.naive_bayes import MultinomialNB

#Algoritma Random Forest
from sklearn.ensemble import RandomForestClassifier

#Algoritma Decision Tree
from sklearn.tree import DecisionTreeClassifier
```

```python
x = data_selected_features
y = df_klasifikasi['label']
#Memisihkan data training dan data testing dengan perbandingan 75:25
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state=6)
```

```python
#print
print('Banyak data x_train :',len(x_train))
print('Banyak data x_test  :',len(x_test))
print('Banyak data y_train :',len(y_train))
print('Banyak data y_test  :',len(y_test))
```

```python
#Training Model

from datetime import datetime
start_time = datetime.now()

#algoritme fitting

#text_algorithm = MultinomialNB()
text_algorithm = svm.SVC(kernel="rbf", C=1.0)
#text_algorithm = RandomForestClassifier()
#text_algorithm = DecisionTreeClassifier()
model = text_algorithm.fit(x_train, y_train)


end_time = datetime.now()
result_time  =end_time-start_time
print("Duration:",result_time)
```

```python
#Prediksi
predicted = model.predict(x_test)
predicted
```

# __5. Model Evaluation__

```python
#Library evaluasi
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

#Mengkomputasi performa model
NewprediksiBenar = (predicted == y_test).sum()
NewprediksiSalah = (predicted != y_test).sum()
    
print("prediksi benar: ", NewprediksiBenar, " data")
print("prediksi salah: ", NewprediksiSalah, " data")
print("Akurasi Algoritma: ", NewprediksiBenar/(NewprediksiBenar+NewprediksiSalah)*100,"%")

# Menggunakan Confusion Matrix untuk Model Evaluasi
CM = confusion_matrix(y_test,predicted)

TN = CM[0][0]
FN = CM[1][0]
TP = CM[1][1]
FP = CM[0][1]
precision    = TP/(TP+FP)
recall       = TP/(TP+FN)
f1 = 2*recall*precision/(recall+precision)

print ("TRUE NEGATIVE (TN):",TN)
print ("FALSE NEGATIVE (FN):",FN)
print ("TRUE POSITIVE (TP):",TP)
print ("FALSE POSITIVE (FP):",FP)
print ("PRECISION:",precision*100,"%")
print ("RECALL:",recall*100,"%")
print ("F1:",f1*100,"%")

print(classification_report(y_test,predicted))
```

```python
#Menggunakan Cross Validation untuk memvalidasi data

from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv        = ShuffleSplit(n_splits=10, test_size=0.25, random_state=6)
accuracy  = (cross_val_score(model,x,y,cv=cv,scoring='accuracy'))
score_accuracy = np.mean(accuracy)
print ("accuracy: ",accuracy)
print ("accuracy: ",score_accuracy)
```
